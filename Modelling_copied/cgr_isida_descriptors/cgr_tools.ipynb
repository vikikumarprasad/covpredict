```python
import os 
import pathlib

from CGRtools import SMILESRead, RDFRead, RDFWrite
import CIMtools
from CIMtools.preprocessing import Fragmentor, CGR
```

Set up path for Fragmentor

```python
path = pathlib.Path(CIMtools.__file__).parent.parent / "Fragmentor"
os.environ['PATH'] += ';' + str(path)
```

Read mapped reaction SMILES and display example reaction

```python
with SMILESRead("to_cgr_mapped_corrected.smi") as file:
    cgr_reactions = file.read()
```

```python
reaction = cgr_reactions[0]
reaction.thiele()
reaction.clean2d()
reaction.standardize()
reaction
```

Standardize and aromatize

```python
for reaction in cgr_reactions:
    reaction.thiele()
    reaction.standardize()
```

Make the sequence-based ISIDA descriptors

```python
cgr = CGR()
```

```python
frg = Fragmentor(fragment_type=3, min_length=2, max_length=6, cgr_dynbonds=1, doallways=True, useformalcharge=True)
X = cgr.transform(cgr_reactions)
X = frg.fit_transform(X)
X.to_csv("isida_sequence_6.csv", index=False)
```

```python
frg = Fragmentor(fragment_type=3, min_length=2, max_length=8, cgr_dynbonds=1, doallways=True, useformalcharge=True)
X = cgr.transform(cgr_reactions)
X = frg.fit_transform(X)
X.to_csv("isida_sequence_8.csv", index=False)
```

Make atom-centered ISIDA descriptors

```python
frg = Fragmentor(fragment_type=6, min_length=2, max_length=4, cgr_dynbonds=1, useformalcharge=True)
X = cgr.transform(cgr_reactions)
X = frg.fit_transform(X)
X.to_csv("isida_atom_4.csv", index=False)
```
